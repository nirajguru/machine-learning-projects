```python
# Importing the necessary libraries
import numpy as np
import pandas as pd
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import BaggingClassifier, AdaBoostClassifier
from xgboost import XGBClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
```

```python
data = load_breast_cancer()
print(load_breast_cancer().data)
print(load_breast_cancer().feature_names)
df = pd.DataFrame(data.data, columns=data.feature_names)
df['target'] = data.target
df
```

```python
X = df.drop(columns=["target"])
y= df["target"]
print(y)
X
```

X has 30 columns instead of 31 columns

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=5)
print(f"Training shape: {X_train.shape}")
print(f"Test shape: {X_test.shape}")
```

## Bagging

```python
# Step 4: Initialize the bagging classifier
base_estimator = DecisionTreeClassifier(random_state=8)
bagging_clf = BaggingClassifier(base_estimator=base_estimator,
                                 random_state=8)
bagging_clf
```

```python
## step 5. Define hyperparameters to tune
param_grid = {
    'n_estimators': [10, 50, 100],
    'max_samples': [0.5, 1.0],
    'max_features': [0.5, 1.0],
    'bootstrap': [True, False],
    'bootstrap_features': [True, False]
}

# Parameters of BaggingClassifier:
# - base_estimator: The base estimator to fit on random subsets of the dataset.
# - n_estimators: The number of base estimators in the ensemble.
# - max_samples: The number of samples to draw from X to train each base estimator.
# - max_features: The number of features to draw from X to train each base estimator.
# - bootstrap: Whether samples are drawn with replacement.
# - bootstrap_features: Whether features are drawn with replacement.
# - random_state: Controls the random resampling of the original dataset.
```

```python
## step 6. set up GridSeach CV
grid_search = GridSearchCV(estimator=bagging_clf,
                           param_grid=param_grid,
                           cv=5,
                           n_jobs=1,
                           scoring="accuracy")
# we are using Grid search (trying to get the best possible combination of hyperparameters)
# with CV (ensuring a robust set of training and test data is used to avoid overfitting and underfitting)
```

```python
## Seto 7: fit the model using GridSearchCV
grid_search.fit(X_train, y_train)
```

```python
# step 8: retrieve the best parameters and estimator
best_parameter = grid_search.best_params_
best_estimator = grid_search.best_estimator_
print(f"Best hyperparameters found: {best_parameter}")
```

```python
# Step 9 . Make predictions
y_pred = best_estimator.predict(X_test)
```

```python
# Step 10. Evaluating the classifier
accuracy = accuracy_score(y_test, y_pred)
print(f"accuracy is {accuracy:.2f}")

# print("classification_report")
print(classification_report(y_test, y_pred))

print("Confusion Matrix***")
print(confusion_matrix(y_test, y_pred))
```

## Boosting

```python
# Initialize boosting
ada_clf = AdaBoostClassifier(base_estimator=base_estimator, random_state=8)
#here base estimator is decision tree
```

```python
param_grid_boosting = {
    "n_estimators": [50, 100, 200],
    "learning_rate": [0.01, 0.1, 1.0]
}
```

```python
grid_search_boosting = GridSearchCV(estimator=ada_clf, 
                                    param_grid=param_grid_boosting, 
                                    cv=5, 
                                    n_jobs=-1,
                                    scoring= "accuracy"
                                    )
```

```python
grid_search_boosting.fit(X_train, y_train)
```

```python
best_params_boosting = grid_search_boosting.best_params_
best_estimator_boosting = grid_search_boosting.best_estimator_
print(f" best hyperparameters for AdaBoost: {best_params_boosting}")
```

```python
# make predictions
y_pred_boosting = best_estimator_boosting.predict(X_test)
```

```python
accuracy_boosting =  accuracy_score(y_test, y_pred_boosting)
print(f"Adaboost accuracy: {accuracy_boosting:.2f}")
```

```python
# Display the classification report for AdaBoost
print("AdaBoost Classification Report:")
print(classification_report(y_test, y_pred_boosting))

# Display the confusion matrix for AdaBoost
print("AdaBoost Confusion Matrix:")
print(confusion_matrix(y_test, y_pred_boosting))
```
